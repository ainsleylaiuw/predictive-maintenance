# MLSMOTE

Multi-Label SMOTE

This notebook is to test the following SMOTE method for oversampling in a multi-label context.

https://medium.com/thecyphy/handling-data-imbalance-in-multi-label-classification-mlsmote-531155416b87

```python
from mlsmote import create_dataset, get_minority_instace, MLSMOTE
```

```python
X, y = create_dataset()                     #Creating a Dataframe
print(f'X Shape: {X.shape}')
print(f'y Shape: {y.shape}')
X
```

```python
y.head()
```

```python
X_sub, y_sub = get_minority_instace(X, y)   #Getting minority instance of that datframe
print(X_sub.shape)
X_sub.head()
```

```python
print(y_sub.shape)
y_sub.head()
```

```python
X_res,y_res = MLSMOTE(X_sub, y_sub, 100)     #Applying MLSMOTE to augment the dataframe

print(f'X_res Shape: {X_res.shape}')
print(f'y_res Shape: {y_res.shape}')
X_res.head()
```

```python
y_res.sum()
```

```python
y.sum()
```
